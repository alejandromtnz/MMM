## Modelo

### Variables de inversin

```python
# Definir las columnas de inversin
investment_columns = [
    'MMM05_INV_CINE', 'MMM05_INV_EXTERIOR', 'MMM05_INV_INTERNET',
    'MMM05_INV_PRENSA', 'MMM05_INV_PRODUCCION', 'MMM05_INV_RADIO',
    'MMM05_INV_REVISTAS', 'MMM05_INV_PlataformasVideo', 'MMM05_INV_VARIOS', 'MMM05_INV_INV_Total'
]

# Asegurarnos de que las variables estn en el DataFrame
df[investment_columns].head()  # Ver las primeras filas
```

### Lag

```python
# Crear lags para las variables de inversin y sustituir NaN por 0
for col in investment_columns:
    df[f'{col}_lag1'] = df[col].shift(1)  # Retardo de 1 periodo
    df[f'{col}_lag2'] = df[col].shift(2)  # Retardo de 2 periodos
    
    # Reemplazar NaN por 0
    df[f'{col}_lag1'] = df[f'{col}_lag1'].fillna(0)
    df[f'{col}_lag2'] = df[f'{col}_lag2'].fillna(0)
    
    print(f"'{col}_lag1', '{col}_lag2', ")
```

### Agrupar las variables de inversin

```python
# Agrupar las variables relevantes
df['Inversion_Cine'] = df['MMM05_INV_CINE'] 
df['Inversion_Plataformas'] = df['MMM05_INV_PlataformasVideo']
df['Inversion_Internet'] = df['MMM05_INV_INTERNET'] + df['MMM05_INV_EXTERIOR']
df['Inversion_Tradicional'] = df[['MMM05_INV_PRENSA', 'MMM05_INV_PRODUCCION', 'MMM05_INV_RADIO', 'MMM05_INV_REVISTAS']].sum(axis=1)
```

### Efecto Adstock

```python
def apply_adstock(x, rate=0.5):
    adstocked = []
    prev = 0
    for val in x:
        curr = val + prev * rate
        adstocked.append(curr)
        prev = curr
    return adstocked

# Aplicar adstock a las variables de inversin
df['Inversion_Cine_adstock'] = apply_adstock(df['Inversion_Cine'], rate=0.6)
df['Inversion_Plataformas_adstock'] = apply_adstock(df['Inversion_Plataformas'], rate=0.6)
df['Inversion_Internet_adstock'] = apply_adstock(df['Inversion_Internet'], rate=0.6)
df['Inversion_Tradicional_adstock'] = apply_adstock(df['Inversion_Tradicional'], rate=0.6)
```

```python
# Definir las variables predictoras (con lags y adstock)
X = df[['Inversion_Cine_adstock', 'Inversion_Plataformas_adstock', 'Inversion_Internet_adstock', 'Inversion_Tradicional_adstock', 
        'MMM04_TIME_working_days', 'Navidad', 'Black_Friday', 'MMM03_OFFLINE_Ticket_medio']]

# Definir el target (ventas o trfico)
y = df['Sales_per_Visit']

# Modelar con regresin lineal (puedes probar otros modelos)
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(X, y)

# Predicciones
predicciones = modelo.predict(X)

# Evaluacin del modelo
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y, predicciones)
r2 = r2_score(y, predicciones)

print(f'MSE: {mse}')
print(f'R: {r2}')
```

### Optimizacin

```python
import numpy as np
from scipy.optimize import minimize

# Definir la funcin objetivo para optimizar las inversiones
def objetivo(inversiones_opt):
    # Restaurar las inversiones optimizadas en el DataFrame (solo para las variables de adstock)
    df['Inversion_Cine_adstock'] = inversiones_opt[0]
    df['Inversion_Plataformas_adstock'] = inversiones_opt[1]
    df['Inversion_Internet_adstock'] = inversiones_opt[2]
    df['Inversion_Tradicional_adstock'] = inversiones_opt[3]
    
    # Actualizar las inversiones con los lags (esto puede estar ya hecho previamente en el DataFrame)
    for col in investment_columns:
        df[f'{col}_lag1'] = df[col].shift(1)  # Retardo de 1 periodo
        df[f'{col}_lag2'] = df[col].shift(2)  # Retardo de 2 periodos
        
        # Reemplazar NaN por 0
        df[f'{col}_lag1'] = df[f'{col}_lag1'].fillna(0)
        df[f'{col}_lag2'] = df[f'{col}_lag2'].fillna(0)

    # Regenerar las variables predictoras para las inversiones optimizadas
    X_opt = df[['Inversion_Cine_adstock', 'Inversion_Plataformas_adstock', 'Inversion_Internet_adstock', 
            'Inversion_Tradicional_adstock', 'MMM04_TIME_working_days', 'Navidad', 'Black_Friday', 'Semana_Santa',
            'MMM03_OFFLINE_Ticket_medio', 'days_since_start', 
                'MMM05_INV_CINE_lag1', 'MMM05_INV_CINE_lag2', 
                'MMM05_INV_EXTERIOR_lag1', 'MMM05_INV_EXTERIOR_lag2', 
                'MMM05_INV_INTERNET_lag1', 'MMM05_INV_INTERNET_lag2', 
                'MMM05_INV_PRENSA_lag1', 'MMM05_INV_PRENSA_lag2', 
                'MMM05_INV_PRODUCCION_lag1', 'MMM05_INV_PRODUCCION_lag2', 
                'MMM05_INV_RADIO_lag1', 'MMM05_INV_RADIO_lag2', 
                'MMM05_INV_REVISTAS_lag1', 'MMM05_INV_REVISTAS_lag2', 
                'MMM05_INV_PlataformasVideo_lag1', 'MMM05_INV_PlataformasVideo_lag2', 
                'MMM05_INV_VARIOS_lag1', 'MMM05_INV_VARIOS_lag2', 
                'MMM05_INV_INV_Total_lag1', 'MMM05_INV_INV_Total_lag2']]

    # Prediccin de ventas para las inversiones optimizadas
    predicciones_opt = modelo.predict(X_opt)

    # Queremos maximizar las ventas, as que devolvemos el negativo de la prediccin
    return -np.mean(predicciones_opt)

# Funcin para ajustar las inversiones para que no superen el mximo
def ajustar_inversiones(inversiones_opt):
    total_inversion = np.sum(inversiones_opt)
    max_inversion = df['MMM05_INV_INV_Total'].max()  # Valor mximo disponible para inversin total
    
    if total_inversion > max_inversion:
        # Ajustar proporcionalmente las inversiones
        ratio = max_inversion / total_inversion
        inversiones_opt = [inv * ratio for inv in inversiones_opt]
    return inversiones_opt

# Establecer lmites para las variables de inversin
limites_inversiones = [(df['Inversion_Cine_adstock'].min(), df['Inversion_Cine_adstock'].max()),
                       (df['Inversion_Plataformas_adstock'].min(), df['Inversion_Plataformas_adstock'].max()),
                       (df['Inversion_Internet_adstock'].min(), df['Inversion_Internet_adstock'].max()),
                       (df['Inversion_Tradicional_adstock'].min(), df['Inversion_Tradicional_adstock'].max())]

# Inicializar las inversiones con los valores medios
inversiones_iniciales = [
    df['Inversion_Cine_adstock'].mean(),
    df['Inversion_Plataformas_adstock'].mean(),
    df['Inversion_Internet_adstock'].mean(),
    df['Inversion_Tradicional_adstock'].mean()
]

# Realizar la optimizacin
resultado_opt = minimize(
    objetivo, 
    inversiones_iniciales, 
    bounds=limites_inversiones, 
    method='L-BFGS-B'
)

# Ajustar las inversiones para que no superen el mximo
inversiones_ajustadas = ajustar_inversiones(resultado_opt.x)
```

```python
# Ver el resultado de la optimizacin
if resultado_opt.success:
    inversiones_optimas = resultado_opt.x

    # Calculamos el total de las inversiones iniciales para obtener el porcentaje utilizado
    total_inversion_inicial = sum(inversiones_iniciales)
    
    # Imprimir resultados truncados a 2 decimales y formateados con comas
    print(f'Las inversiones ptimas para maximizar las ventas son:')
    
    # Definir las inversiones y su nombre
    nombres_inversiones = ['Cine', 'Plataformas', 'Internet', 'Tradicional']
    
    for i, nombre in enumerate(nombres_inversiones):
        # Formatear la inversin truncada a dos decimales y con coma para miles
        inversion_formateada = f"{inversiones_optimas[i]:,.2f}"
        
        # Calcular el porcentaje de la inversin utilizada
        porcentaje_utilizado = (inversiones_optimas[i] / total_inversion_inicial) * 100
        
        # Imprimir la inversin con el porcentaje de inversin utilizado
        print(f'{nombre}: {inversion_formateada} ({porcentaje_utilizado:.2f}%)')
else:
    print("La optimizacin no fue exitosa. Intenta ajustar los parmetros.")
```
