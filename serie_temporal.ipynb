```python
import pandas as pd
import pyodbc
import logging
import os
from functools import reduce

# Configura logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Configuracin de conexin
AZURE_SERVER = 'uaxmathfis.database.windows.net'
AZURE_DATABASE = 'usecases'
AZURE_DRIVER = '{ODBC Driver 17 for SQL Server}'
CSV_PATH = 'datos_combinados.csv'

def get_azure_connection():
    """Establece conexin con Azure SQL usando autenticacin interactiva"""
    conn_str = f"DRIVER={AZURE_DRIVER};SERVER={AZURE_SERVER};DATABASE={AZURE_DATABASE};Authentication=ActiveDirectoryInteractive"
    try:
        conn = pyodbc.connect(conn_str)
        logging.info("Conexin exitosa a Azure SQL")
        return conn
    except Exception as e:
        logging.error(f"Error de conexin: {e}")
        return None

def download_and_combine_tables():
    """Descarga todas las tablas y las combina correctamente por fecha"""
    # Eliminar el archivo CSV si existe
    if os.path.exists(CSV_PATH):
        try:
            os.remove(CSV_PATH)
            logging.info(f"Archivo existente {CSV_PATH} eliminado")
        except Exception as e:
            logging.error(f"No se pudo eliminar {CSV_PATH}: {e}")

    queries = {
        'MMM01_WEB': "SELECT * FROM DATAEX.MMM01_WEB",
        'MMM02_VISIT': "SELECT * FROM DATAEX.MMM02_VISIT",
        'MMM03_OFFLINE': "SELECT * FROM DATAEX.MMM03_OFFLINE",
        'MMM04_TIME': "SELECT * FROM DATAEX.MMM04_TIME",
        'MMM05_INV': "SELECT * FROM DATAEX.MMM05_INV"
    }

    conn = get_azure_connection()
    if not conn:
        return pd.DataFrame()

    try:
        # Descargar todas las tablas
        dfs = {}
        for name, query in queries.items():
            try:
                df = pd.read_sql(query, conn)
                if 'ID_Date' in df.columns:
                    df['ID_Date'] = pd.to_datetime(df['ID_Date'].astype(str), format=('%Y%m'))
                    df = df.drop_duplicates(subset=['ID_Date'])
                    # Agregamos el nombre de la tabla como prefijo a las columnas
                    df.columns = [f'{name}_{col}' if col != 'ID_Date' else col for col in df.columns]
                    dfs[name] = df.set_index('ID_Date')
                    logging.info(f"Tabla {name} descargada ({len(df)} registros)")
                else:
                    logging.warning(f"Tabla {name} no tiene columna ID_Date")
            except Exception as e:
                logging.error(f"Error procesando {name}: {e}")

        # Combinar horizontalmente por fecha
        if dfs:
            # Fusionar todas las tablas usando merge secuencial
            combined = reduce(lambda left, right: pd.merge(
                left, right, 
                left_index=True, right_index=True, 
                how='outer'
            ), dfs.values())
            
            # Guardar a CSV (se crear nuevo archivo)
            combined.to_csv(CSV_PATH)
            logging.info(f"Nuevo archivo {CSV_PATH} creado ({len(combined)} registros combinados)")
            return combined
        else:
            logging.warning("No se descargaron datos vlidos")
            return pd.DataFrame()

    finally:
        conn.close()

def load_from_csv():
    """Carga los datos previamente guardados desde el CSV"""
    try:
        if os.path.exists(CSV_PATH):
            return pd.read_csv(CSV_PATH, parse_dates=['ID_Date'], index_col='ID_Date')
        else:
            logging.warning(f"Archivo {CSV_PATH} no encontrado")
            return pd.DataFrame()
    except Exception as e:
        logging.error(f"Error cargando CSV: {e}")
        return pd.DataFrame()

# Ejecutar la descarga (siempre crear nuevo archivo)
df = download_and_combine_tables()

# Para cargar los datos posteriormente (si el archivo existe)
# df = load_from_csv()
```

```python
# 2. Anlisis Exploratorio y Preparacin de datos
def preparar_datos(df):
    # Verificar datos faltantes
    print("Datos faltantes por columna:")
    print(df.isnull().sum())
    
    # Imputar datos faltantes si es necesario 
    df_limpio = df.fillna(method='ffill').fillna(method='bfill')
    
    # Verificar y eliminar duplicados
    duplicados = df_limpio.index.duplicated().sum()
    print(f"Filas duplicadas: {duplicados}")
    df_limpio = df_limpio[~df_limpio.index.duplicated()]
    
    # Asegurarse que la serie est ordenada por fecha
    df_limpio = df_limpio.sort_index()
    
    # Convertir frecuencia si es necesario 
    # df_limpio = df_limpio.asfreq('MS')
    
    return df_limpio
```

```python
# 3. Descomposicin de la serie temporal
def descomponer_serie(df, columna_objetivo, periodo=12):
    """
    Descompone la serie temporal en tendencia, estacionalidad y residuos
    """
    # Asegurarse que no hay valores nulos
    serie = df[columna_objetivo].fillna(method='ffill')
    
    # Descomposicin
    descomposicion = seasonal_decompose(serie, model='additive', period=periodo)
    
    # Crear DataFrame con los componentes
    componentes = pd.DataFrame({
        'observado': serie,
        'tendencia': descomposicion.trend,
        'estacionalidad': descomposicion.seasonal,
        'residuos': descomposicion.resid
    })
    
    # Calcular media mvil
    componentes['media_movil'] = serie.rolling(window=periodo).mean()
    
    # Visualizar componentes
    fig, axes = plt.subplots(5, 1, figsize=(14, 16))
    componentes['observado'].plot(ax=axes[0], title='Serie Observada')
    componentes['tendencia'].plot(ax=axes[1], title='Tendencia')
    componentes['estacionalidad'].plot(ax=axes[2], title='Estacionalidad')
    componentes['residuos'].plot(ax=axes[3], title='Residuos')
    componentes['media_movil'].plot(ax=axes[4], title=f'Media Mvil (ventana={periodo})')
    plt.tight_layout()
    plt.show()
    
    return componentes
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

def descomponer_serie(df, columna_objetivo, periodo=None):
    """
    Descompone la serie temporal en tendencia, estacionalidad y residuos.
    - df: DataFrame con los datos.
    - columna_objetivo: Nombre de la columna numrica a analizar.
    - periodo: (Opcional) Ciclo estacional. Si es None, no se calcula estacionalidad.
    """
    # Limpieza de datos (rellena NaN con el ltimo valor vlido)
    serie = df[columna_objetivo].fillna(method='ffill')
    
    # Descomposicin (solo tendencia si periodo=None)
    descomposicion = seasonal_decompose(
        serie, 
        model='additive', 
        period=periodo,  # Si es None, solo calcula tendencia y residuos
        extrapolate_trend='freq'  # Maneja bordes
    )
    
    # DataFrame con componentes
    componentes = pd.DataFrame({
        'observado': serie,
        'tendencia': descomposicion.trend,
        'residuos': descomposicion.resid
    })
    
    # Aadir estacionalidad solo si se especific periodo
    if periodo is not None:
        componentes['estacionalidad'] = descomposicion.seasonal
    
    # Grficos
    num_plots = 4 if periodo is not None else 3
    fig, axes = plt.subplots(num_plots, 1, figsize=(14, 3*num_plots))
    
    axes[0].plot(componentes['observado'], label='Observado')
    axes[0].set_title('Serie Original')
    
    axes[1].plot(componentes['tendencia'], label='Tendencia', color='orange')
    axes[1].set_title('Tendencia')
    
    if periodo is not None:
        axes[2].plot(componentes['estacionalidad'], label='Estacionalidad', color='green')
        axes[2].set_title(f'Estacionalidad (perodo={periodo})')
        axes[3].plot(componentes['residuos'], label='Residuos', color='red')
        axes[3].set_title('Residuos')
    else:
        axes[2].plot(componentes['residuos'], label='Residuos', color='red')
        axes[2].set_title('Residuos')
    
    plt.tight_layout()
    plt.show()
    
    return componentes
```

```python
# 4. Modelo SARIMA con auto_arima y variables exgenas
def modelar_sarima_exogenas(df, columna_objetivo, columnas_exogenas):
    """
    Ajusta un modelo SARIMAX usando auto_arima para determinar los parmetros ptimos
    e incluye variables exgenas con transformacin polinmica
    """
    # Preparar datos
    y = df[columna_objetivo].dropna()
    
    # Preparar variables exgenas con transformaciones polinmicas (grado 0 a 4)
    X = df[columnas_exogenas].dropna()
    X_poly = X.copy()
    
    # Aadir transformaciones polinmicas (x^2, x^3, x^4)
    for col in columnas_exogenas:
        for grado in range(2, 5):  # Polinomios de grado 2 a 4
            X_poly[f'{col}_grado_{grado}'] = X[col] ** grado
    
    # Alinear ndices entre y y X_poly
    indices_comunes = y.index.intersection(X_poly.index)
    y = y.loc[indices_comunes]
    X_poly = X_poly.loc[indices_comunes]
    
    # Encontrar mejores parmetros con auto_arima
    print("Ejecutando auto_arima para encontrar mejores parmetros...")
    modelo_auto = auto_arima(
        y,
        exogenous=X_poly,
        start_p=0, start_q=0,
        max_p=4, max_q=4,
        start_P=0, start_Q=0,
        max_P=2, max_Q=2,
        d=None, D=None, max_d=2, max_D=1,
        seasonal=True, m=periodo,
        trace=True,
        error_action='ignore',
        suppress_warnings=True,
        stepwise=True
    )
    
    print(f"Mejores parmetros encontrados: {modelo_auto.order}, {modelo_auto.seasonal_order}")
    
    # Ajustar el modelo SARIMAX con los parmetros encontrados
    p, d, q = modelo_auto.order
    P, D, Q, m = modelo_auto.seasonal_order
    
    modelo_final = SARIMAX(
        y,
        exog=X_poly,
        order=(p, d, q),
        seasonal_order=(P, D, Q, m),
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    
    resultados = modelo_final.fit(disp=False)
    print(resultados.summary())
    
    # Evaluar el modelo
    y_pred = resultados.predict(exog=X_poly)
    mse = ((y - y_pred) ** 2).mean()
    print(f"MSE: {mse}")
    
    # Visualizar predicciones vs valores reales
    plt.figure(figsize=(12, 6))
    plt.plot(y, label='Observado')
    plt.plot(y_pred, label='Predicho', color='red')
    plt.legend()
    plt.title('Valores Observados vs Predichos')
    plt.show()
    
    # Diagnstico de residuos
    residuos = resultados.resid
    fig, axes = plt.subplots(2, 2, figsize=(14, 8))
    
    # Residuos vs tiempo
    residuos.plot(ax=axes[0, 0], title='Residuos')
    axes[0, 0].axhline(y=0, color='r', linestyle='-')
    
    # Histograma de residuos
    sns.histplot(residuos, kde=True, ax=axes[0, 1])
    axes[0, 1].set_title('Distribucin de Residuos')
    
    # QQ Plot
    from scipy import stats
    stats.probplot(residuos, dist="norm", plot=axes[1, 0])
    axes[1, 0].set_title('QQ Plot')
    
    # Autocorrelacin de residuos
    from statsmodels.graphics.tsaplots import plot_acf
    plot_acf(residuos, ax=axes[1, 1], title='Autocorrelacin de Residuos')
    
    plt.tight_layout()
    plt.show()
    
    return resultados, X_poly
```

```python
# 5. Funcin para hacer predicciones futuras
def predecir_futuro(modelo, X_futuro, pasos=12):
    """
    Realiza predicciones para perodos futuros
    """
    # Generar predicciones
    predicciones = modelo.get_forecast(steps=pasos, exog=X_futuro)
    
    # Intervalos de confianza
    pred_ci = predicciones.conf_int()
    
    # Crear DataFrame con predicciones
    pred_mean = predicciones.predicted_mean
    df_predicciones = pd.DataFrame({
        'prediccion': pred_mean,
        'limite_inferior': pred_ci.iloc[:, 0],
        'limite_superior': pred_ci.iloc[:, 1]
    })
    
    # Visualizar predicciones
    plt.figure(figsize=(12, 6))
    plt.plot(df_predicciones.index, df_predicciones['prediccion'], label='Prediccin')
    plt.fill_between(
        df_predicciones.index,
        df_predicciones['limite_inferior'],
        df_predicciones['limite_superior'],
        color='pink', alpha=0.3,
        label='Intervalo de Confianza (95%)'
    )
    plt.legend()
    plt.title('Predicciones Futuras')
    plt.show()
    
    return df_predicciones
```
